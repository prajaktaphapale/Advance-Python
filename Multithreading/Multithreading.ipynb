## Multithreading

Main goal of multithreading is to save time to execute code by using multiple threads at a time.

```python
import time
```

```python
time.sleep(3)
print ("Hi All")
```

```python
def hypotenuse(a, b):
    print("hypotenuse function started")
    time.sleep(3)
    h= (a**2 + b**2)**(1/2)
    print(f"hypotenuse is {h:.4f}")
    return h
```

```python
hypotenuse(3, 4)
```

```python
def simple_int(P, N, R):
    print(f"Simple_int function sgtarted")
    time.sleep(2)
    I = (P*N*R)/100
    print(f"I={I}")
    return I    
```

```python
I = simple_int(P=50000, N=4, R=7.1)
print(I)
```

```python
import time
```

Single thread code always exceutes in a sequence
Multithreading for differnt functions

```python
from threading import Thread
```

```python
%%time
# Create threads
th1 = Thread(target=hypotenuse, args= (3, 4))
th2 = Thread(target= simple_int, args=(50000, 4, 7))

# Strat all threads
th1.start()
th2.start()

# Wait for all threads to finish

th1.join()
th2.join()
```

## Multithreading for single function but multipe values

```python
def square(n):
    time.sleep(3)
    s= n**2
    print(f"Square of given number {n} is {s}")
    return s
```

```python
square(5)
```

```python
%%time
square(1.4)
```

```python
num = [2, 3, 4, 5]
num
```

```python
len(num)
```

```python
%%time
for i in num:
    square(i)
```

## multithreading provide thread for each number in lis

```python
num
```

```python
%%time
# Create a thread for each other
threads = []
for i in num:
    th = Thread(target=square, args=(i,))
    th.start()
    threads.append(th)

# Wait for all threads to finish
for j in threads:
    j.join()
```
