### List Vs Array Multiplication of Numbers

```python
b= [3, 4, 5, 6, 7]
type(b)
```

```python
b*3
```

### Get Product of all elements inside list

```python
t = [2, 3, 4, 5, 6]
type(t)
```

```python
p = 1
for i in t:
    p = p*i
```

```python
p
```

```python
np.prod(t)
```

```python
int(np.prod(t))
```
